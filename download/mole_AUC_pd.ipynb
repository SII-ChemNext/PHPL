```python
import pickle
from chembl_webresource_client.new_client import new_client
import pandas as pd
with open("activity_AUC-pd.pkl","rb") as f:
    aucs = pickle.load(f)
molecule_chembl_ids = aucs["molecule_chembl_id"]
molecule_chembl_ids = list(molecule_chembl_ids)
print(len(molecule_chembl_ids))
```

```python
molecule = new_client.molecule
mols = molecule.filter(molecule_chembl_id__in=molecule_chembl_ids).only(['molecule_chembl_id', 'molecule_structures'])
mols = pd.DataFrame(mols)
print(len(mols))
with open("mole_AUC_pd.pkl","wb") as f:
    pickle.dump(mols,f)
```
