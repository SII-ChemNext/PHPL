```python
import pickle
import os
from tqdm import tqdm
paths = ['./mol/','./mol-2/','./mol-3/','./mol-4/']
Allmolecule = []
for path in paths:
    files = []
    filess = os.listdir(path)
    for file in filess:
        if len(file.split(".")) == 2:
            if file.split(".")[1] == "pkl":
                files.append(file)
    len(files)
    try:
        for file in tqdm(files):
            file_path = str(path) + str(file)
            with open(file_path,"rb") as f:
                pkl = pickle.load(f)
                Allmolecule.append(pkl)
    except:
        print(file)
print(len(Allmolecule))
with open("Allmolecule.pkl","wb") as f:
    pickle.dump(Allmolecule,f)
```
