```python
import pickle
import os
from tqdm import tqdm
paths = ['./IC50-3/']
IC50 = []
for path in paths:
    files = []
    filess = os.listdir(path)
    for file in filess:
        if len(file.split(".")) == 2:
            if file.split(".")[1] == "pkl":
                files.append(file)
    len(files)
    for file in tqdm(files):
        file_path = str(path) + str(file)
        try:
            with open(file_path,"rb") as f:
                pkl = pickle.load(f)
                IC50.append(pkl)
        except:
            print(file)
print(len(IC50))
with open("IC50-3.pkl","wb") as f:
    pickle.dump(IC50,f)
```
