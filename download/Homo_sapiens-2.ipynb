```python
import pickle
import os
from tqdm import tqdm
paths = ['./Homo_sapiens/']
Homo_sapiens = []
for path in paths:
    files = []
    filess = os.listdir(path)
    for file in filess:
        if len(file.split(".")) == 2:
            if file.split(".")[1] == "pkl":
                files.append(file)
    len(files)
    for file in tqdm(files):
        file_path = str(path) + str(file)
        try:
            with open(file_path,"rb") as f:
                pkl = pickle.load(f)
                Homo_sapiens.append(pkl)
        except:
            print(file)
print(len(Homo_sapiens))
with open("Homo_sapiens.pkl","wb") as f:
    pickle.dump(Homo_sapiens,f)
```
