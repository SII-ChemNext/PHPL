```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
import pickle
acti = new_client.activity

```

```python
acti = acti.filter(standard_type__exact='Tmax')
```

```python
len(acti)
for i in range(len(acti)):
    try:
        filename = './Tmax/' + str(i) + '.pkl'
        with open(filename,"wb") as f:
            pickle.dump(acti[i],f)
    except:
        pass
```

```python
import os
path_error = "./Vdss-error"
errors_all = os.listdir(path_error)
errors = []
for error in errors_all:
    if error.split(".")[1] == "pkl":
        errors.append(error.split(".")[0])

for i in errors:
    try:
        filename = './Vdss-error/' + i + '.pkl'
        with open(filename,"wb") as f:
            i = int(i)
            pickle.dump(acti[i],f)
    except:
        pass
```

```python
import pickle
import os
from tqdm import tqdm
paths = ['./Cmax/']
Cmax = []
for path in paths:
    files = []
    filess = os.listdir(path)
    for file in filess:
        if len(file.split(".")) == 2:
            if file.split(".")[1] == "pkl":
                files.append(file)
    len(files)
    for file in tqdm(files):
        file_path = str(path) + str(file)
        try:
            with open(file_path,"rb") as f:
                pkl = pickle.load(f)
                Cmax.append(pkl)
        except:
            print(file)
print(len(Cmax))
with open("Cmax.pkl","wb") as f:
    pickle.dump(Cmax,f)
```
