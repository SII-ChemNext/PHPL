```python
import pickle
import pandas as pd
from chembl_webresource_client.new_client import new_client

assay = new_client.assay
result = assay.filter(assay_organism__iexact="Homo sapiens").only(['assay_chembl_id','assay_organism','target_chembl_id'])
len(result)
```

```python
for i in range(0,len(result)):
    filename = './Homo_sapiens/' + str(i) + '.pkl'
    with open(filename,"wb") as f:
        pickle.dump(result[i],f)
```
