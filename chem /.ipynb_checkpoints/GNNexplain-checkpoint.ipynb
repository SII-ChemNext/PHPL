```python
from rdkit import Chem
from rdkit.Chem import Draw

def visualize_topk(smiles, atom_score, k=5, out_path=None):
    """
    用 RDKit 绘制 top-k 原子重要性图
    """

    mol = Chem.MolFromSmiles(smiles)
    num_atoms = mol.GetNumAtoms()

    # 获取 top-k 原子
    topk_idx, topk_val = get_topk_atoms(atom_score, k=k)

    # 归一化（用于颜色渐变）
    scores = atom_score.cpu().numpy()
    max_s = scores.max()
    norm = scores / max_s if max_s > 0 else scores

    # 颜色设置：Top-k 原子最红
    atom_colors = {}
    for i in range(num_atoms):
        if i in topk_idx:
            # 最重要 → 纯红
            atom_colors[i] = (1.0, 0.0, 0.0)
        else:
            # 其他：红色程度按重要性递减
            r = float(norm[i])
            atom_colors[i] = (r, 0.7*(1-r), 0.7*(1-r))   # 渐变：红→粉→白

    img = Draw.MolToImage(
        mol,
        size=(450, 320),
        highlightAtoms = topk_idx,
        highlightAtomColors=atom_colors,
        highlightAtomRadii={i: 0.45 for i in range(num_atoms)},
    )

    if out_path:
        img.save(out_path)
        print(f"[Saved] Top-k explanation saved to {out_path}")

    return img
```

```python
def get_topk_atoms(atom_score, k=5):
    """
    返回 top-k 原子的索引（按重要性排序）
    """
    atom_score = atom_score.detach().cpu()
    topk = atom_score.topk(k)
    top_indices = topk.indices.tolist()
    top_values = topk.values.tolist()
    return top_indices, top_values
```

```python
############# IMPORTS #############
import argparse
import torch
import torch.nn as nn
from torch_geometric.data import DataLoader
import torch.nn.functional as F

from model import GNN_graphpred
from loader_desc import MoleculeDataset
from tqdm import tqdm
from captum.attr import IntegratedGradients

import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Draw



############################################################
# 1. Batch-Compatible Wrapper for IG on desc only
############################################################

class AtomIGWrapper(nn.Module):
    """
    全部逻辑写在这里！
    IG 输入 h0_batch → wrapper 里做 message passing → pooling → desc → pred
    """

    def __init__(self, model, data):
        super().__init__()
        self.model = model
        self.data = data

        # 图结构
        self.edge_index = data.edge_index
        self.edge_attr = data.edge_attr

        # batch（单图=全0）
        if getattr(data, "batch", None) is None:
            self.batch = torch.zeros(data.x.size(0), dtype=torch.long, device=data.x.device)
        else:
            self.batch = data.batch

        # 图级描述符
        self.desc = data.desc.unsqueeze(0)   # shape [1, desc_dim]

    def forward(self, h0_batch):
        """
        h0_batch: [N_steps, num_nodes, hidden_dim]
        输出: [N_steps, num_tasks]
        """
        outputs = []

        for i in range(h0_batch.size(0)):

            h = h0_batch[i]          # 取 IG 插值后的 h0
            h_list = [h]             # 对应你 GNN 的 h_list

            # ===== 1. 模拟 GNN 的 forward（跳过 embedding）=====
            for layer in range(self.model.gnn.num_layer):
                h = self.model.gnn.gnns[layer](h_list[layer], self.edge_index, self.edge_attr)
                h = self.model.gnn.batch_norms[layer](h)

                if layer == self.model.gnn.num_layer - 1:
                    h = F.dropout(h, self.model.gnn.drop_ratio, training=False)
                else:
                    h = F.dropout(F.relu(h), self.model.gnn.drop_ratio, training=False)

                h_list.append(h)

            # ===== 2. JK 汇聚 =====
            JK = self.model.JK
            if JK == "concat":
                node_rep = torch.cat(h_list, dim=1)
            elif JK == "last":
                node_rep = h_list[-1]
            elif JK == "max":
                node_rep = torch.max(torch.stack(h_list), dim=0)[0]
            elif JK == "sum":
                node_rep = torch.stack(h_list, dim=0).sum(dim=0)

            # ===== 3. graphpred 中的后处理 =====
            node_rep = self.model.node_layerNorm(node_rep)
            node_rep = F.dropout(node_rep, self.model.drop_ratio, training=False)

            # pooling
            graph_rep = self.model.pool(node_rep, self.batch)

            # desc
            if self.model.desc_emb > 0:
                d = self.desc
                d = self.model.desc_mlp(d)
                d = self.model.desc_bn(d)
                graph_rep = torch.cat([graph_rep, d], dim=1)

            # ===== 4. 最终预测 =====
            out = self.model.graph_pred_linear(graph_rep.float())
            outputs.append(out)

        return torch.cat(outputs, dim=0)




############################################################
# 2. IG only for desc
############################################################

def explain_desc_with_ig(model, data, target_index=0, steps=200):
    model.eval()
    device = next(model.parameters()).device
    data = data.to(device)
    h0 = model.gnn.motif_embedding1(data.x[:, 0])+model.gnn.motif_embedding2(data.x[:, 1])

    wrapper = DescIGBatchWrapper(model, data).to(device)
    ig = IntegratedGradients(wrapper)

    desc = data.desc.unsqueeze(0).to(device)      # [1,216]
    baseline = torch.zeros_like(desc)             # baseline: 全0

    attributions = ig.attribute(
        inputs=desc,
        baselines=baseline,
        target=target_index,
        n_steps=steps
    )

    return attributions.squeeze(0)  # [216]



############################################################
# 3. 显示最重要的 desc 特征
############################################################

def print_topk_desc_importance(desc_attr, topk=10):
    vals = desc_attr.detach().cpu().numpy()
    abs_vals = np.abs(vals)
    idx = abs_vals.argsort()[::-1]

    print(f"\nTop {topk} 描述符维度（按绝对重要性排序）:")
    for i in idx[:topk]:
        print(f"  - 维度 {i}: 贡献={vals[i]:.4f}, |贡献|={abs_vals[i]:.4f}")



############################################################
# 4. 主程序（可直接运行）
############################################################

if __name__ == "__main__":
    parser = argparse.ArgumentParser(description='IG for desc explanation')

    parser.add_argument('--device', type=int, default=0)
    parser.add_argument('--batch_size', type=int, default=1)
    parser.add_argument('--num_layer', type=int, default=5)
    parser.add_argument('--emb_dim', type=int, default=300)
    parser.add_argument('--dropout_ratio', type=float, default=0.0)
    parser.add_argument('--graph_pooling', type=str, default="mean")
    parser.add_argument('--JK', type=str, default="last")
    parser.add_argument('--filename', type=str,
        default='/inspire/hdd/project/materialscienceresearch/wenyihao-240208090182/PEMAL-master2/results/20250723/final_0_desc_mae_0.0_1/lr_0.0001_decay_1e-05_bz_16_drop_0.1_temp_1.0/best_model.pth')
    parser.add_argument('--num_workers', type=int, default=4)
    parser.add_argument('--gnn_type', type=str, default="gin")

    args = parser.parse_args(args=[])

    device = torch.device(f"cuda:{args.device}" if torch.cuda.is_available() else "cpu")

    # 加载验证集
    valid_dataset_name = "4_valid"
    valid_dataset = MoleculeDataset(
        "../dataset_new_desc/" + valid_dataset_name,
        dataset=valid_dataset_name
    )
    val_loader = DataLoader(
        valid_dataset, batch_size=args.batch_size,
        shuffle=False, num_workers=args.num_workers
    )

    # smiles
    df = pd.read_csv('../dataset_new_desc/4_valid/processed/smiles.csv', header=None)
    smiles_list = df[0].tolist()

    # 加载模型
    num_tasks = 3
    model = GNN_graphpred(
        args.num_layer, args.emb_dim, num_tasks,
        JK=args.JK,
        drop_ratio=args.dropout_ratio,
        graph_pooling=args.graph_pooling,
        gnn_type=args.gnn_type,
        classnum=1,
        desc_emb=216,
    )

    print("Loading pretrained model:", args.filename)
    checkpoint = torch.load(args.filename, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device)
    model.eval()


    #############################################
    # 解释一个样本
    #############################################

    for idx, data_i in enumerate(tqdm(val_loader, desc="Iteration")):
        data_i = data_i.to(device)            # Data or Batch（但 batch_size=1）

        smiles_i = smiles_list[idx]
        print(f"Now explaining sample #{idx}, SMILES = {smiles_i}")

        print("desc shape:", data_i.desc.shape)

        # === IG explaining ===
        # 1. 得到 h0（embedding 输出）
        with torch.no_grad():
            h0 = model.gnn.motif_embedding1(data_i.x[:,0]) + model.gnn.motif_embedding2(data_i.x[:,1])
        h0 = h0.unsqueeze(0)

        baseline = torch.zeros_like(h0)

        # 2. Wrapper
        wrapper = AtomIGWrapper(model, data_i)
        ig = IntegratedGradients(wrapper)

        # 3. IG 解释
        attr = ig.attribute(inputs=h0, baselines=baseline, target=2)

        # 4. 原子贡献
        atom_score = attr.squeeze(0).abs().sum(1)
        
        # print("每个原子的贡献值（从 IG 得到）:")
        # for i, s in enumerate(atom_score.tolist()):
        #     print(f"Atom {i}: score = {s:.4f}")

        # 你已经得到 atom_score, smiles_i
        # img = visualize_topk(smiles_i, atom_score, k=5, out_path=f"../explain/t1_2{idx}topk_explanation.png")
        # img
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem.Draw import rdMolDraw2D
import matplotlib.pyplot as plt

def mol_to_heatmap_svg(mol, atom_score, cmap_name="coolwarm",
                       width=300, height=240, bond_px=20):
    """
    用 RDKit 的 SVG 渲染每个原子不同颜色的热力图。
    返回 svg 字符串，可直接保存或在 notebook 中展示。
    """

    # 1. 归一化热力值
    scores = atom_score.detach().cpu().numpy().astype(float)
    min_s, max_s = scores.min(), scores.max()
    if max_s > min_s:
        norm = (scores - min_s) / (max_s - min_s)
    else:
        norm = scores
    norm = np.clip(norm, 0, 1)

    # # gamma 增强（可选）
    # gamma = 2.2
    # norm = norm ** gamma

    # colormap（matplotlib）
    cmap = plt.colormaps.get_cmap(cmap_name)

    # 2. 生成 RDKit 颜色
    atom_colors = {}
    for i, v in enumerate(norm):
        r, g, b, a = cmap(float(v))
        atom_colors[i] = (float(r), float(g), float(b))
    
    # bond score
    bond_score = []
    bond_colors = {}
    cmap = plt.colormaps.get_cmap("coolwarm")

    for bond in mol.GetBonds():
        i = bond.GetBeginAtomIdx()
        j = bond.GetEndAtomIdx()
        s = atom_score[i] + atom_score[j]   # 最常用的 bond attribution
        bond_score.append(s)

    # normalize
    bond_score = bond_score.detach().cpu().numpy().astype(float)
    bs_min, bs_max = bond_score.min(), bond_score.max()
    bond_norm = (bond_score - bs_min) / (bs_max - bs_min)

    for k, v in enumerate(bond_norm):
        r,g,b,a = cmap(float(v))
        bond_colors[k] = (float(r), float(g), float(b))

    # 3. 开始绘制 SVG
    d = rdMolDraw2D.MolDraw2DSVG(width, height)
    opts = d.drawOptions()

    # 让原子不显示数字
    opts.addAtomIndices = False

    # 统一键长度
    if hasattr(opts, "fixedBondLength"):
        opts.fixedBondLength = bond_px

    # 4. 调用 SVG 绘制
    d.DrawMolecule(
        mol,
        highlightAtoms=list(range(mol.GetNumAtoms())),
        highlightAtomColors=atom_colors,
        highlightAtomRadii={i:0.45 for i in range(mol.GetNumAtoms())},
        highlightBonds=[],
    )
    d.FinishDrawing()

    svg = d.GetDrawingText()
    return svg

mol = Chem.MolFromSmiles(smiles_list[idx])
svg = mol_to_heatmap_svg(mol, atom_score, cmap_name="coolwarm")

# # 保存 SVG
# with open("heatmap.svg", "w") as f:
#     f.write(svg)

# 若在 notebook 中展示：
from IPython.display import SVG
SVG(svg)
```

```python
from rdkit.Chem.Draw import rdMolDraw2D
from PIL import Image
import io
import numpy as np

def mol_to_png(mol, atom_colors, width=900, height=720, bond_px=45):
    d = rdMolDraw2D.MolDraw2DCairo(width, height)
    opts = d.drawOptions()
    opts.addAtomIndices = False
    if hasattr(opts, "fixedBondLength"):
        opts.fixedBondLength = bond_px

    d.DrawMolecule(
        mol,
        highlightAtoms=list(range(mol.GetNumAtoms())),
        highlightAtomColors=atom_colors,
        highlightAtomRadii={i:0.45 for i in atom_colors.keys()},
        highlightBondColors=bond_colors
    )
    d.FinishDrawing()
    png = d.GetDrawingText()

    return Image.open(io.BytesIO(png))

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors

def draw_molecule_with_bar(mol, atom_score, cmap_name="coolwarm",
                           out_path="mol_with_bar.png"):

    # 归一化
    scores = atom_score.detach().cpu().numpy().astype(float)
    norm = (scores - scores.min()) / (scores.max() - scores.min())
    cmap = plt.colormaps.get_cmap(cmap_name)

    # RDKit 颜色
    atom_colors = {}
    for i,v in enumerate(norm):
        r,g,b,a = cmap(float(v))
        atom_colors[i] = (float(r), float(g), float(b))

    # RDKit 画 molecule PNG
    mol_img = mol_to_png(mol, atom_colors)

    # Matplotlib 画 molecule + colorbar
    fig, ax = plt.subplots(figsize=(6,4), dpi=150)
    ax.imshow(mol_img)
    ax.axis("off")

    # 横向 colorbar
    cax = fig.add_axes([0.15, 0.05, 0.7, 0.1])  # 位置：左、下、宽、高
    norm_obj = colors.Normalize(vmin=scores.min(), vmax=scores.max())
    cb = plt.colorbar(cm.ScalarMappable(norm=norm_obj, cmap=cmap_name),
                      cax=cax, orientation="horizontal")
    cb.set_label("Atom attribution", fontsize=10)

    fig.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"[Saved] {out_path}")

mol = Chem.MolFromSmiles(smiles_list[idx])
draw_molecule_with_bar(mol, atom_score, out_path="heatmap_with_bar.png")
```
